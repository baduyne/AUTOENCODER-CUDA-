```python
import pandas as pd
import matplotlib.pyplot as plt

train = pd.read_csv("train_log.csv")

plt.figure(figsize=(10, 5))
plt.plot(train["train_loss"])
plt.title("Train Loss per Eval Step")
plt.xlabel("Eval Step")
plt.ylabel("Loss")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

eval_df = pd.read_csv("eval_log.csv")

plt.figure(figsize=(10, 5))
plt.plot(eval_df["epoch"], eval_df["eval_loss"], marker='o')
plt.title("Eval Loss per Epoch")
plt.xlabel("Epoch")
plt.ylabel("Eval Loss")
plt.grid(True)
plt.show()
```

```python

# Tnh tng thi gian 1 epoch (ms)
epoch_times = train.groupby("epoch")["step_time_ms"].sum()

print("=== Tng thi gian mi epoch (ms) ===")
print(epoch_times)
print("\n")

# Tnh trung bnh thi gian cho mi checkpoint (50 step)
avg_50step = train["step_time_ms"].mean()

print(f"Thi gian trung bnh cho mi 50 step: {avg_50step:.3f} ms")

# V y l thi gian  chy 1 block 50 step,
# suy ra thi gian 1 step:
avg_1step = avg_50step / 50

print(f"Thi gian trung bnh cho 1 step: {avg_1step:.6f} ms/step")
```
