```python
import pandas as pd
import matplotlib.pyplot as plt

train = pd.read_csv("train_log.csv")

plt.figure(figsize=(10, 5))
plt.plot(train["train_loss"])
plt.title("Train Loss per Step")
plt.xlabel("Step")
plt.ylabel("Loss")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

eval_df = pd.read_csv("eval_log.csv")

plt.figure(figsize=(10, 5))
plt.plot(eval_df["epoch"], eval_df["eval_loss"], marker='o')
plt.title("Eval Loss per Epoch")
plt.xlabel("Epoch")
plt.ylabel("Eval Loss")
plt.grid(True)
plt.show()
```
